```python
import pdfplumber
import pandas as pd
import re
import os

def extract_fault_cases_pdf(pdf_path):
    print(f"ğŸ“– ë‘ ë²ˆì§¸ PDF ì²˜ë¦¬ ì‹œì‘: {os.path.basename(pdf_path)}")
    
    extracted_data = []
    
    with pdfplumber.open(pdf_path) as pdf:
        for page_num, page in enumerate(pdf.pages):
            print(f"   ğŸ“„ í˜ì´ì§€ {page_num + 1} ì²˜ë¦¬ ì¤‘...")
            
            tables = page.extract_tables()
            for table in tables:
                for row in table:
                    if row and len(row) >= 6:  # í‘œ êµ¬ì¡°ë¥¼ ê¸°ì¤€ìœ¼ë¡œ í•„í„°ë§
                        # ì •ì œëœ ê°’ ì¶”ì¶œ
                        code = extract_code(row[0])
                        name = clean_text(row[1])
                        base_ratio = clean_text(row[2])
                        mod_factor = clean_text(row[3])
                        final_ratio = clean_text(row[4])
                        case_number = clean_text(row[5])

                        if code or name:
                            extracted_data.append({
                                "source_pdf": "ê³¼ì‹¤ë¹„ìœ¨ì‹¬ì˜ì‚¬ë¡€",
                                "page_number": page_num + 1,
                                "accident_code": code,
                                "accident_name": name,
                                "basic_fault_ratio": base_ratio,
                                "modification_factors": mod_factor,
                                "related_cases": case_number,
                                "video_link": "",
                                "raw_text": ' | '.join([clean_text(c) for c in row if c])
                            })
    
    print(f"âœ… ë‘ ë²ˆì§¸ PDF ì™„ë£Œ: ì´ {len(extracted_data)}ê°œ ì‚¬ë¡€ ì¶”ì¶œ")
    return extracted_data

def extract_code(text):
    """ì‚¬ê³ ìœ í˜• ì½”ë“œ ì¶”ì¶œ"""
    if not text:
        return ""
    match = re.search(r'([ë³´ì°¨ì´pmPM]\d+)', text, re.IGNORECASE)
    return match.group(1).upper() if match else ""

def clean_text(text):
    """í…ìŠ¤íŠ¸ ì „ì²˜ë¦¬"""
    if not text:
        return ""
    return re.sub(r'\s+', ' ', text.strip())

def validate_and_save_data(data_list, filename):
    """ë°ì´í„° ê²€ì¦ ë° ì €ì¥"""
    if not data_list:
        print("âŒ ì¶”ì¶œëœ ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤.")
        return
    
    df = pd.DataFrame(data_list)
    df.to_csv(filename, index=False, encoding='utf-8-sig')
    print(f"ğŸ’¾ CSV íŒŒì¼ ì €ì¥ ì™„ë£Œ: {filename}")

# === ì‹¤í–‰ ===
if __name__ == "__main__":
    pdf_path_2 = r"C:\project\2stProject_jun\jun\ê³¼ì‹¤ë¹„ìœ¨PDF\(ìµœì¢…)ê³¼ì‹¤ë¹„ìœ¨ì‹¬ì˜ì‚¬ë¡€_(54MB).pdf"
    print("ğŸš€ Step 2 ì‹¬ì˜ì‚¬ë¡€ PDF ì²˜ë¦¬ ì‹œì‘")
    data_cases = extract_fault_cases_pdf(pdf_path_2)
    
    validate_and_save_data(data_cases, "temp_step2_cases_final.csv")
```

# 1. ìˆ˜ì •ë³´ì™„(EasyOCR)

```python
import pdfplumber
import pandas as pd
import os
import easyocr
import numpy as np
from PIL import Image
import time
import re

class SimpleTextExtractor:
    def __init__(self, use_gpu=True):
        """
        ë‹¨ìˆœí•œ í…ìŠ¤íŠ¸ ì¶”ì¶œê¸° - ë³µì¡í•œ í•„í„°ë§ ì—†ì´ ëª¨ë“  í…ìŠ¤íŠ¸ ì €ì¥
        """
        print("ğŸš€ ë‹¨ìˆœ í…ìŠ¤íŠ¸ ì¶”ì¶œê¸° ì´ˆê¸°í™” ì¤‘...")
        
        try:
            self.reader = easyocr.Reader(['ko', 'en'], gpu=use_gpu)
            print(f"âœ… EasyOCR ì´ˆê¸°í™” ì™„ë£Œ (GPU: {use_gpu})")
        except Exception as e:
            print(f"âš ï¸  GPU ì´ˆê¸°í™” ì‹¤íŒ¨, CPU ëª¨ë“œë¡œ ì „í™˜: {e}")
            self.reader = easyocr.Reader(['ko', 'en'], gpu=False)

    def extract_all_text(self, pdf_path, output_format='both', max_pages=None):
        """
        ëª¨ë“  í˜ì´ì§€ì—ì„œ í…ìŠ¤íŠ¸ ì¶”ì¶œí•˜ì—¬ ì €ì¥
        
        Args:
            pdf_path: PDF íŒŒì¼ ê²½ë¡œ
            output_format: 'csv', 'txt', 'both' ì¤‘ ì„ íƒ
            max_pages: ì²˜ë¦¬í•  ìµœëŒ€ í˜ì´ì§€ ìˆ˜ (Noneì´ë©´ ì „ì²´)
        """
        print(f"ğŸ“– ì „ì²´ í…ìŠ¤íŠ¸ ì¶”ì¶œ ì‹œì‘: {os.path.basename(pdf_path)}")
        
        results = []
        
        with pdfplumber.open(pdf_path) as pdf:
            total_pages = len(pdf.pages)
            if max_pages:
                total_pages = min(total_pages, max_pages)
            
            print(f"ğŸ“„ ì´ {total_pages}í˜ì´ì§€ ì²˜ë¦¬ ì˜ˆì •")
            
            for page_num in range(total_pages):
                page = pdf.pages[page_num]
                print(f"   ğŸ“„ í˜ì´ì§€ {page_num + 1}/{total_pages} ì²˜ë¦¬ ì¤‘...")
                
                # EasyOCR ì¶”ì¶œ
                text_ocr = self.extract_with_easyocr(page, page_num + 1)
                
                # pdfplumber ì¶”ì¶œ
                text_pdf = self.extract_with_pdfplumber(page)
                
                # ìµœì  í…ìŠ¤íŠ¸ ì„ íƒ
                best_text, method = self.choose_better_text(text_ocr, text_pdf)
                
                # ê²°ê³¼ ì €ì¥
                page_result = {
                    'page_number': page_num + 1,
                    'extraction_method': method,
                    'text_length': len(best_text),
                    'ocr_text': text_ocr,
                    'pdf_text': text_pdf,
                    'selected_text': best_text,
                    'has_korean': len(re.findall(r'[ê°€-í£]', best_text)) > 0,
                    'has_numbers': len(re.findall(r'\d', best_text)) > 0,
                    'contains_keywords': self.check_keywords(best_text)
                }
                
                results.append(page_result)
                
                print(f"      âœ… ì™„ë£Œ - {method}: {len(best_text)}ì")
        
        # ê²°ê³¼ ì €ì¥
        base_filename = os.path.splitext(os.path.basename(pdf_path))[0]
        
        if output_format in ['csv', 'both']:
            self.save_to_csv(results, f"{base_filename}_extracted_text.csv")
        
        if output_format in ['txt', 'both']:
            self.save_to_txt(results, f"{base_filename}_extracted_text.txt")
        
        # í†µê³„ ì¶œë ¥
        self.print_statistics(results)
        
        return results

    def extract_with_easyocr(self, page, page_num):
        """EasyOCRë¡œ í…ìŠ¤íŠ¸ ì¶”ì¶œ"""
        try:
            start_time = time.time()
            
            # ê³ í•´ìƒë„ ì´ë¯¸ì§€ ë³€í™˜
            pil_image = page.to_image(resolution=300).original
            image_array = np.array(pil_image)
            
            # EasyOCR ì‹¤í–‰
            results = self.reader.readtext(image_array)
            
            if results:
                # ìœ„ì¹˜ ê¸°ë°˜ ì •ë ¬ (ìœ„â†’ì•„ë˜, ì™¼ìª½â†’ì˜¤ë¥¸ìª½)
                sorted_results = sorted(results, key=lambda x: (x[0][0][1], x[0][0][0]))
                extracted_text = '\n'.join([result[1] for result in sorted_results])
                
                elapsed = time.time() - start_time
                print(f"      ğŸ” EasyOCR: {len(extracted_text)}ì ({elapsed:.1f}ì´ˆ)")
                return extracted_text
            else:
                print(f"      âŒ EasyOCR: í…ìŠ¤íŠ¸ ì—†ìŒ")
                return ""
                
        except Exception as e:
            print(f"      âŒ EasyOCR ì˜¤ë¥˜: {e}")
            return ""

    def extract_with_pdfplumber(self, page):
        """pdfplumberë¡œ í…ìŠ¤íŠ¸ ì¶”ì¶œ"""
        try:
            text = page.extract_text()
            if text:
                print(f"      ğŸ“„ pdfplumber: {len(text)}ì")
                return text
            else:
                print(f"      âŒ pdfplumber: í…ìŠ¤íŠ¸ ì—†ìŒ")
                return ""
        except Exception as e:
            print(f"      âŒ pdfplumber ì˜¤ë¥˜: {e}")
            return ""

    def choose_better_text(self, text_ocr, text_pdf):
        """ë” ë‚˜ì€ í…ìŠ¤íŠ¸ ì„ íƒ"""
        
        # ë‘˜ ë‹¤ ë¹„ì–´ìˆìœ¼ë©´
        if not text_ocr and not text_pdf:
            return "", "None"
        
        # í•˜ë‚˜ë§Œ ìˆìœ¼ë©´ ê·¸ê²ƒ ì„ íƒ
        if not text_ocr:
            return text_pdf, "pdfplumber"
        if not text_pdf:
            return text_ocr, "EasyOCR"
        
        # ë‘˜ ë‹¤ ìˆìœ¼ë©´ í’ˆì§ˆ ë¹„êµ
        score_ocr = self.calculate_simple_score(text_ocr)
        score_pdf = self.calculate_simple_score(text_pdf)
        
        # ê¸¸ì´ë„ ê³ ë ¤ (ë„ˆë¬´ ì§§ìœ¼ë©´ ì ìˆ˜ ê°ì )
        if len(text_ocr) < 50:
            score_ocr *= 0.5
        if len(text_pdf) < 50:
            score_pdf *= 0.5
        
        if score_ocr > score_pdf:
            return text_ocr, "EasyOCR"
        else:
            return text_pdf, "pdfplumber"

    def calculate_simple_score(self, text):
        """ê°„ë‹¨í•œ í…ìŠ¤íŠ¸ í’ˆì§ˆ ì ìˆ˜"""
        if not text:
            return 0
        
        score = 0
        
        # ê¸°ë³¸ ì ìˆ˜: í…ìŠ¤íŠ¸ ê¸¸ì´
        score += len(text) / 100
        
        # í•œêµ­ì–´ ìˆìœ¼ë©´ ë³´ë„ˆìŠ¤
        korean_count = len(re.findall(r'[ê°€-í£]', text))
        score += korean_count / 10
        
        # ìˆ«ì ìˆìœ¼ë©´ ë³´ë„ˆìŠ¤
        number_count = len(re.findall(r'\d', text))
        score += number_count / 20
        
        # íŠ¹ìˆ˜ í‚¤ì›Œë“œ ë³´ë„ˆìŠ¤
        keywords = ['ë³´1', 'ë³´2', 'ì°¨1', 'ì°¨2', 'ê³¼ì‹¤', 'ë¹„ìœ¨', '%', 'ëŒ€ë²•ì›', 'íŒê²°']
        for keyword in keywords:
            if keyword in text:
                score += 5
        
        return score

    def check_keywords(self, text):
        """ì£¼ìš” í‚¤ì›Œë“œ í¬í•¨ ì—¬ë¶€ í™•ì¸"""
        keywords = [
            'ê³¼ì‹¤', 'ë¹„ìœ¨', 'ì‚¬ê³ ', 'ë³´í–‰ì', 'ìë™ì°¨', 'êµì°¨ë¡œ', 
            'ì§ì§„', 'ì¢ŒíšŒì „', 'ìš°íšŒì „', 'ì‹ í˜¸', 'íš¡ë‹¨ë³´ë„',
            'ë³´1', 'ë³´2', 'ë³´3', 'ì°¨1', 'ì°¨2', 'ì°¨3',
            'ëŒ€ë²•ì›', 'íŒê²°', 'â‘ ', 'â‘¡', 'â‘¢'
        ]
        
        found_keywords = [kw for kw in keywords if kw in text]
        return ', '.join(found_keywords) if found_keywords else 'None'

    def save_to_csv(self, results, filename):
        """CSV íŒŒì¼ë¡œ ì €ì¥"""
        print(f"\nğŸ’¾ CSV ì €ì¥ ì¤‘: {filename}")
        
        # DataFrame ìƒì„±
        df_data = []
        for result in results:
            df_data.append({
                'page_number': result['page_number'],
                'extraction_method': result['extraction_method'],
                'text_length': result['text_length'],
                'has_korean': result['has_korean'],
                'has_numbers': result['has_numbers'],
                'contains_keywords': result['contains_keywords'],
                'selected_text': result['selected_text'],
                'ocr_text': result['ocr_text'],
                'pdf_text': result['pdf_text']
            })
        
        df = pd.DataFrame(df_data)
        df.to_csv(filename, index=False, encoding='utf-8-sig')
        print(f"âœ… CSV ì €ì¥ ì™„ë£Œ: {filename}")

    def save_to_txt(self, results, filename):
        """TXT íŒŒì¼ë¡œ ì €ì¥ (ì½ê¸° í¸í•œ í˜•íƒœ)"""
        print(f"\nğŸ’¾ TXT ì €ì¥ ì¤‘: {filename}")
        
        with open(filename, 'w', encoding='utf-8') as f:
            f.write("=" * 80 + "\n")
            f.write("PDF í…ìŠ¤íŠ¸ ì¶”ì¶œ ê²°ê³¼\n")
            f.write("=" * 80 + "\n\n")
            
            for result in results:
                f.write(f"ğŸ“„ í˜ì´ì§€ {result['page_number']}\n")
                f.write(f"ğŸ“Š ì¶”ì¶œë°©ë²•: {result['extraction_method']}\n")
                f.write(f"ğŸ“ í…ìŠ¤íŠ¸ ê¸¸ì´: {result['text_length']}ì\n")
                f.write(f"ğŸ” í¬í•¨ í‚¤ì›Œë“œ: {result['contains_keywords']}\n")
                f.write("-" * 60 + "\n")
                f.write(result['selected_text'])
                f.write("\n\n" + "=" * 80 + "\n\n")
        
        print(f"âœ… TXT ì €ì¥ ì™„ë£Œ: {filename}")

    def print_statistics(self, results):
        """ì¶”ì¶œ í†µê³„ ì¶œë ¥"""
        total_pages = len(results)
        
        # ì¶”ì¶œ ë°©ë²•ë³„ í†µê³„
        method_counts = {}
        for result in results:
            method = result['extraction_method']
            method_counts[method] = method_counts.get(method, 0) + 1
        
        # í…ìŠ¤íŠ¸ í’ˆì§ˆ í†µê³„
        text_lengths = [r['text_length'] for r in results]
        avg_length = sum(text_lengths) / len(text_lengths) if text_lengths else 0
        
        korean_pages = sum(1 for r in results if r['has_korean'])
        keyword_pages = sum(1 for r in results if r['contains_keywords'] != 'None')
        
        print(f"\nğŸ“Š ìµœì¢… ì¶”ì¶œ í†µê³„")
        print("=" * 50)
        print(f"ğŸ“„ ì´ í˜ì´ì§€: {total_pages}")
        print(f"ğŸ“ í‰ê·  í…ìŠ¤íŠ¸ ê¸¸ì´: {avg_length:.0f}ì")
        print(f"ğŸ‡°ğŸ‡· í•œêµ­ì–´ í¬í•¨ í˜ì´ì§€: {korean_pages}í˜ì´ì§€ ({korean_pages/total_pages*100:.1f}%)")
        print(f"ğŸ” í‚¤ì›Œë“œ í¬í•¨ í˜ì´ì§€: {keyword_pages}í˜ì´ì§€ ({keyword_pages/total_pages*100:.1f}%)")
        
        print(f"\nğŸ† ì¶”ì¶œ ë°©ë²•ë³„ ì„ íƒ íšŸìˆ˜:")
        for method, count in method_counts.items():
            print(f"   {method}: {count}íšŒ ({count/total_pages*100:.1f}%)")
        
        # ìµœê³  í’ˆì§ˆ í˜ì´ì§€ Top 5
        top_pages = sorted(results, key=lambda x: x['text_length'], reverse=True)[:5]
        print(f"\nğŸ… í…ìŠ¤íŠ¸ ê¸¸ì´ ìƒìœ„ 5í˜ì´ì§€:")
        for i, page in enumerate(top_pages):
            keywords = page['contains_keywords'][:50] + "..." if len(page['contains_keywords']) > 50 else page['contains_keywords']
            print(f"   {i+1}. í˜ì´ì§€ {page['page_number']}: {page['text_length']}ì ({page['extraction_method']}) - {keywords}")

def main():
    """ë©”ì¸ ì‹¤í–‰ í•¨ìˆ˜"""
    print("ğŸš€ ë‹¨ìˆœ í…ìŠ¤íŠ¸ ì¶”ì¶œê¸° ì‹œì‘")
    print("="*60)
    
    # PDF íŒŒì¼ ê²½ë¡œ
    pdf_path = r"C:\project\2stProject_jun\jun\ê³¼ì‹¤ë¹„ìœ¨PDF\(ìµœì¢…)ê³¼ì‹¤ë¹„ìœ¨ì‹¬ì˜ì‚¬ë¡€_(54MB).pdf"
    
    if not os.path.exists(pdf_path):
        print(f"âŒ PDF íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {pdf_path}")
        return
    
    # ì¶”ì¶œê¸° ì´ˆê¸°í™”
    extractor = SimpleTextExtractor(use_gpu=True)
    
    # ì„ íƒ ë©”ë‰´
    print("\nğŸ“‹ ì¶”ì¶œ ì˜µì…˜ì„ ì„ íƒí•˜ì„¸ìš”:")
    print("1. ì „ì²´ í˜ì´ì§€ ì¶”ì¶œ (CSV + TXT)")
    print("2. ì²˜ìŒ 20í˜ì´ì§€ë§Œ í…ŒìŠ¤íŠ¸ (CSV + TXT)")
    print("3. CSVë§Œ ì €ì¥")
    print("4. TXTë§Œ ì €ì¥")
    
    choice = input("\nì„ íƒ (1-4): ").strip()
    
    if choice == '1':
        results = extractor.extract_all_text(pdf_path, output_format='both')
    elif choice == '2':
        results = extractor.extract_all_text(pdf_path, output_format='both', max_pages=20)
    elif choice == '3':
        results = extractor.extract_all_text(pdf_path, output_format='csv')
    elif choice == '4':
        results = extractor.extract_all_text(pdf_path, output_format='txt')
    else:
        print("ê¸°ë³¸ê°’ìœ¼ë¡œ ì²˜ìŒ 20í˜ì´ì§€ í…ŒìŠ¤íŠ¸ë¥¼ ì‹¤í–‰í•©ë‹ˆë‹¤.")
        results = extractor.extract_all_text(pdf_path, output_format='both', max_pages=20)
    
    print(f"\nğŸ‰ ì¶”ì¶œ ì™„ë£Œ!")
    print(f"ğŸ“ ìƒì„±ëœ íŒŒì¼ë“¤ì„ í™•ì¸í•˜ì„¸ìš”.")

if __name__ == "__main__":
    main()
    
```

# 2. ìˆ˜ì •ë³´ì™„(ë§ˆí¬ë‹¤ìš´)

```python
# 1ë‹¨ê³„: í•„ìˆ˜ íŒ¨í‚¤ì§€ ì„¤ì¹˜
import subprocess
import sys

def install_package(package):
    subprocess.check_call([sys.executable, "-m", "pip", "install", package])

print("í•„ìˆ˜ íŒ¨í‚¤ì§€ ì„¤ì¹˜ ì¤‘...")
try:
    install_package("PyMuPDF")
    install_package("pandas")
    print("ì„¤ì¹˜ ì™„ë£Œ!")
except Exception as e:
    print(f"ì„¤ì¹˜ ì˜¤ë¥˜: {e}")

# 2ë‹¨ê³„: íŒ¨í‚¤ì§€ import ë° PDF ë³€í™˜ í•¨ìˆ˜
try:
    import fitz  # PyMuPDF
    import pandas as pd
    import io
    import os
    import re
    print("íŒ¨í‚¤ì§€ import ì„±ê³µ!")
    
    def extract_text_from_pdf_simple(pdf_path):
        """ê°„ë‹¨í•œ PDF í…ìŠ¤íŠ¸ ì¶”ì¶œ (OCR ì—†ìŒ)"""
        if not os.path.exists(pdf_path):
            return f"íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {pdf_path}"
        
        doc = fitz.open(pdf_path)
        markdown_content = []
        
        for page_num in range(len(doc)):
            page = doc.load_page(page_num)
            
            # í˜ì´ì§€ ì œëª©
            markdown_content.append(f"# í˜ì´ì§€ {page_num + 1}\n")
            
            # í…ìŠ¤íŠ¸ ì¶”ì¶œ
            text = page.get_text()
            if text.strip():
                # ê¸°ë³¸ ì •ë¦¬
                lines = text.split('\n')
                cleaned_lines = []
                
                for line in lines:
                    line = line.strip()
                    if not line:
                        continue
                    
                    # ì œëª© íŒ¨í„´ ê°ì§€
                    if re.match(r'^[\d\.\)]+\s+', line):
                        cleaned_lines.append(f"### {line}")
                    else:
                        cleaned_lines.append(line)
                
                markdown_content.append("\n".join(cleaned_lines) + "\n")
            
            # í‘œ ì¶”ì¶œ ì‹œë„
            blocks = page.get_text("dict")["blocks"]
            for block_idx, block in enumerate(blocks):
                if "lines" in block:
                    lines = []
                    for line in block["lines"]:
                        spans = []
                        for span in line["spans"]:
                            spans.append(span["text"])
                        line_text = " ".join(spans).strip()
                        if line_text:
                            lines.append(line_text)
                    
                    # í‘œ íŒ¨í„´ ê°ì§€
                    if len(lines) > 1:
                        table_data = []
                        for line in lines:
                            cols = re.split(r'\s{3,}|\t', line)
                            if len(cols) > 1:
                                table_data.append(cols)
                        
                        if len(table_data) > 1:
                            markdown_content.append(f"## í‘œ {block_idx + 1}\n")
                            # ë§ˆí¬ë‹¤ìš´ í…Œì´ë¸” ìƒì„±
                            max_cols = max(len(row) for row in table_data)
                            for row in table_data:
                                while len(row) < max_cols:
                                    row.append("")
                            
                            # í—¤ë”
                            header = "| " + " | ".join(table_data[0]) + " |"
                            separator = "| " + " | ".join(["---"] * max_cols) + " |"
                            markdown_content.append(header)
                            markdown_content.append(separator)
                            
                            # ë°ì´í„° í–‰
                            for row in table_data[1:]:
                                data_row = "| " + " | ".join(row) + " |"
                                markdown_content.append(data_row)
                            markdown_content.append("")
        
        doc.close()
        return "\n".join(markdown_content)
    
    # 3ë‹¨ê³„: PDF ë³€í™˜ ì‹¤í–‰
    pdf_path = r"C:\project\2stProject_jun\jun\ê³¼ì‹¤ë¹„ìœ¨PDF\(ìµœì¢…)ê³¼ì‹¤ë¹„ìœ¨ì‹¬ì˜ì‚¬ë¡€_(54MB).pdf"
    
    print("PDF ë³€í™˜ ì‹œì‘...")
    result = extract_text_from_pdf_simple(pdf_path)
    
    if "íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤" in result:
        print(result)
        print("íŒŒì¼ ê²½ë¡œë¥¼ í™•ì¸í•´ì£¼ì„¸ìš”.")
    else:
        # ê²°ê³¼ ì €ì¥
        output_path = pdf_path.replace('.pdf', '.md')
        with open(output_path, 'w', encoding='utf-8') as f:
            f.write(result)
        
        print(f"ë³€í™˜ ì™„ë£Œ: {output_path}")
        print(f"ë³€í™˜ëœ ë‚´ìš© ê¸¸ì´: {len(result)} ë¬¸ì")
        
        # ì²˜ìŒ 500ì ë¯¸ë¦¬ë³´ê¸°
        print("\n=== ë³€í™˜ ê²°ê³¼ ë¯¸ë¦¬ë³´ê¸° ===")
        print(result[:500])
        print("...")

except ImportError as e:
    print(f"import ì˜¤ë¥˜: {e}")
    print("ì»¤ë„ì„ ë‹¤ì‹œ ì‹œì‘í•œ í›„ ë‹¤ì‹œ ì‹¤í–‰í•´ë³´ì„¸ìš”.")
except Exception as e:
    print(f"ì‹¤í–‰ ì˜¤ë¥˜: {e}")
```

```python
# ë³€í™˜ëœ ë§ˆí¬ë‹¤ìš´ íŒŒì¼ ë‚´ìš© í™•ì¸
import os

def analyze_converted_file(md_path):
    """ë³€í™˜ëœ ë§ˆí¬ë‹¤ìš´ íŒŒì¼ ë¶„ì„"""
    
    if not os.path.exists(md_path):
        print(f"íŒŒì¼ì´ ì—†ìŠµë‹ˆë‹¤: {md_path}")
        return
    
    with open(md_path, 'r', encoding='utf-8') as f:
        content = f.read()
    
    lines = content.split('\n')
    
    # ê¸°ë³¸ í†µê³„
    print("=== ë³€í™˜ íŒŒì¼ ë¶„ì„ ===")
    print(f"ì´ ë¬¸ì ìˆ˜: {len(content):,}")
    print(f"ì´ ì¤„ ìˆ˜: {len(lines):,}")
    
    # í˜ì´ì§€ ìˆ˜ ê³„ì‚°
    pages = [line for line in lines if line.startswith('# í˜ì´ì§€')]
    print(f"ì´ í˜ì´ì§€ ìˆ˜: {len(pages)}")
    
    # ì œëª©/êµ¬ì¡° ë¶„ì„
    headers = [line for line in lines if line.startswith('#')]
    print(f"ì œëª© ê°œìˆ˜: {len(headers)}")
    
    # í‘œ ê°œìˆ˜
    tables = [line for line in lines if line.startswith('## í‘œ')]
    print(f"í‘œ ê°œìˆ˜: {len(tables)}")
    
    # ì²˜ìŒ 20ì¤„ ë¯¸ë¦¬ë³´ê¸°
    print("\n=== ì²˜ìŒ 20ì¤„ ë¯¸ë¦¬ë³´ê¸° ===")
    for i, line in enumerate(lines[:20]):
        if line.strip():
            print(f"{i+1:2d}: {line}")
    
    # ëª©ì°¨ ì¶”ì¶œ
    print("\n=== ì£¼ìš” ì œëª© êµ¬ì¡° ===")
    for line in lines[:100]:  # ì²˜ìŒ 100ì¤„ì—ì„œ ì œëª© ì°¾ê¸°
        if line.startswith('### ') and ('.' in line or ')' in line):
            print(line)
    
    return content

# ë³€í™˜ëœ íŒŒì¼ ë¶„ì„
md_path = r"C:\project\2stProject_jun\jun\ê³¼ì‹¤ë¹„ìœ¨PDF\(ìµœì¢…)ê³¼ì‹¤ë¹„ìœ¨ì‹¬ì˜ì‚¬ë¡€_(54MB).md"
content = analyze_converted_file(md_path)

# íŠ¹ì • í˜ì´ì§€ ë‚´ìš© í™•ì¸ (ì˜ˆ: ì²« ë²ˆì§¸ í˜ì´ì§€)
def show_page_content(content, page_num, max_lines=30):
    """íŠ¹ì • í˜ì´ì§€ ë‚´ìš© í‘œì‹œ"""
    lines = content.split('\n')
    page_start = -1
    page_end = -1
    
    for i, line in enumerate(lines):
        if line == f"# í˜ì´ì§€ {page_num}":
            page_start = i
        elif line.startswith(f"# í˜ì´ì§€ {page_num + 1}"):
            page_end = i
            break
    
    if page_start == -1:
        print(f"í˜ì´ì§€ {page_num}ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        return
    
    if page_end == -1:
        page_end = len(lines)
    
    print(f"\n=== í˜ì´ì§€ {page_num} ë‚´ìš© (ìµœëŒ€ {max_lines}ì¤„) ===")
    page_lines = lines[page_start:page_end]
    
    for i, line in enumerate(page_lines[:max_lines]):
        if line.strip():
            print(f"{i+1:2d}: {line}")
    
    if len(page_lines) > max_lines:
        print(f"... (ì´ {len(page_lines)}ì¤„ ì¤‘ {max_lines}ì¤„ë§Œ í‘œì‹œ)")

# ì²« ë²ˆì§¸ í˜ì´ì§€ ë‚´ìš© í™•ì¸
if content:
    show_page_content(content, 1)
```
